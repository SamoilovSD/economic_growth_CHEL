```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
from scipy.optimize import curve_fit
from tqdm import tqdm
import csv

rc = {"font.family" : "serif", 
      "mathtext.fontset" : "stix"}
plt.rcParams.update(rc)
plt.rcParams["font.serif"] = ["Times New Roman"] + plt.rcParams["font.serif"]
matplotlib.rcParams.update({'font.size': 14})
```

```python
def objective(x, a, b, c):
    return a * x + b * x**2 + c
```

```python
mas_GPD = [8.25,4.5,4.5,4.2,4.5,4.5,4.5,3.3,2.7,4.8,3.75,3.5,3,1.8,1.5,-1.5,-4.2,-8.25]
mas_EXP = [15.95,11.7,13,17,20.2,24.4,23.4,29.78,28.72,26.59,29.78,31.9,33,29.7,32,36.17,41.7,43.82]
# China, Thailand, Indonesia, Peru, Mexico,Columbia, Argentina, Turkey, Marocco
# Iran, Egypt, Brazil, Venezuela, S. Africa, Algeria, Romania, Russia, Ukraine
```

```python
popt, _ = curve_fit(objective, mas_EXP, mas_GPD)
# summarize the parameter values
a, b, c = popt
print('y = %.5f * x + %.5f * x^2 + %.5f' % (a, b, c))
```

```python
x = np.linspace(10,45,1000)
y = -0.02035*x**2 + 0.76152*x  -1.56117
```

```python
plt.figure(figsize=(10,6))
plt.plot(mas_EXP, mas_GPD, 'o', label='Анализируемые данные')
plt.plot(x, y, 'r', label='Аппроксимирующая прямая')
plt.plot([0,100],[0,0],c='black')
plt.plot([15,15],[-10,10], c='black',linestyle='--')
plt.plot([23,23],[-10,10], c='black',linestyle='--')
plt.legend()
plt.grid()
plt.text(35,4.5,f'$R^{2} = 0.88 $',horizontalalignment='center', verticalalignment='center')
plt.text(40-2,5.5,f'$y = -0.02x^{2}+ 0.76x - 1.56 $',horizontalalignment='center', verticalalignment='center')
plt.xlim([0, 50])
plt.ylim([-10, 10])
plt.xlabel('Государственные расходы в % ВВП')
plt.ylabel('Темп роста ВВП в %')
plt.savefig('18 стран_исправ',dpi = 300)
```
